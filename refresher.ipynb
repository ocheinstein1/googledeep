```python
#define a function able to find the circumference and area of a circle
```

```python
def getcircle(radius=1):
  pi=3.142
  area= pi*radius*radius
  circumference= 2*pi*radius
  return (area, circumference)
```

```python
getcircle()
```

```python
getcircle(radius=4)
```
